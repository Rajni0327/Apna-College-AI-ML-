# Series

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))


#index
print(s[0])
print(s[2])
```

## important
index means position , and label means , named index

```python
s = pd.Series([23, 45,  67, 34], index = ["Adam", "Eve", "Bob", "Bean"])
print(s)

#excessing labelled based 
print(s["Adam"])

print(s.index)
```

```python
s
```

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s1 + s2)

s1[0] = 100
changed_s1 = s1.drop(0)

print(s1)
print(changed_s1)
```

```python
s1
```
