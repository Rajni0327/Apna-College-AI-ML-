# Transforming Data(Feature Engineering)
- apply()
- map()
- assign()
- replace(old, new)

```python
df
```

```python
df2 = df.copy()
```

```python
df2["tax"] = df2["income"].apply(lambda x : "20%" if  x >= 50000 else "10%")  #apply() is the higher order function

df2
```

```python
#map()

gender_map = {"Male":"M",
             "Female":"F",
             "Unknown":"U"}

df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
df2.assign(new_income = df2["income"] * 1.1)
```

```python
df2["country"] = df2["country"].replace("US", "United States")
df2
```
